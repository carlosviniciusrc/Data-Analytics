```python
# Importanto Bibliotecas 
import pandas as pd
import numpy as np
```

```python
# Criando o dataframe
df = pd.read_csv("dados/dataset.csv")
```

```python
# Visualizando os dados
display(df)
```

```python
# Verificando a quantidade valores vazios
df.isna().sum()
```

Criando a moda para preencher os valores:

```python
# Verificando qual valor se repete mais na coluna de Quantidade
df["Quantidade"].value_counts()
```

```python
# Criando a moda
moda = df["Quantidade"].value_counts().index[0]

print(moda)
```

```python
# Aplicando valor da moda para as linhas com valores ausentes
df["Quantidade"] = df["Quantidade"].fillna(value=moda)

# Exibindo o DF atualizado
display(df)
```

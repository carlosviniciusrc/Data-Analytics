### Construindo Plots

Plot é uma representação gráfica de dados em uma figura. Em outras palavras, é um gráfico que mostra a relação entre duas ou mais variáveis. Um plot pode ser criado usando uma biblioteca de visualização de dados, como o Matplotlib em Python.

Os tipos mais comuns de plots incluem gráficos de linhas, gráficos de dispersão histogramas e gráficos de barras. Cada tipo de plot é adequado para diferentes tipos de dados e propósitos de visualização. Por exemplo, um gráfico de linhas é útil para visualizar tendências em uma série temporal, enquanto um gráfico de dispersão é útil para mostrar a relação entre duas variáveis contínuas.

A criação de um plot envolve a escolha dos dados a serem plotados, a seleção do tipo de plot e a configuração das opções de plotagem, como tamanho da figura, rótulos de eixo, título, cores, entre outros. Um plot bem criado pode fornecer insights valiosos sobre os dados, facilitando a compreensão e a comunicação das informações.

```python
# Importando as bibliotecas
import matplotlib as mpl

# O matplotlib.pyplot é um coleção de funções e estilos do Matplotlib
# O matplotlib.pyplot é o módulo que você mais usa no dia a dia com matplotlib,
# porque ele facilita a criação rápida de gráficos sem precisar mexer diretamente
# nas classes internas de figura e eixos.
import matplotlib.pyplot as plt
```

#### .plot





```python
# Crindo grafico com .plot():
plt.plot([1,3,5], [2,4,7], color='red')

# Vai exibir o gráfico
plt.show()
```

```python
# Criando listas
x = [2,3,5]
y = [3,5,7]
```

```python
# Chamando o Pyplot para criar o grafico com as variaves
plt.plot(x,y)

# Os métodos .xlabel() e .ylabel() do matplotlib.pyplot são usados 
# para adicionar rótulos (nomes) aos eixos de um gráfico.
plt.xlabel('Variável 1')
plt.ylabel('Variável 2')

# O plt.title() é uma função do matplotlib.pyplot usada para adicionar 
# um título ao gráfico.
plt.title('Teste Plot')
```

```python
# Nova lista
x2 = [1,2,3]
y2 = [11,12,15]
```

```python
# label='Gráfico com Matplotlib'

# Esse argumento nomeia a linha desenhada, ou seja, dá um rótulo para a
# legenda do gráfico. Sozinho, não mostra nada. Para aparecer, você precisa chamar
# o método .legend()
plt.plot(x2, y2, label = 'Gráfico com Matplotlib')
plt.legend()
plt.show()
```

```python
# Exemplo Gemini:

# Passo 2: Criar dados de exemplo
# Dados para o eixo X (por exemplo, meses do ano)
meses = [1, 2, 3, 4, 5, 6]
# Dados para o eixo Y (por exemplo, vendas de dois produtos)
vendas_produto_a = [100, 120, 150, 130, 170, 180]
vendas_produto_b = [80, 85, 90, 110, 105, 120]

# Passo 3: Criar o gráfico usando .plot()
# Adicionamos o 'label' aqui para que a legenda funcione
plt.plot(meses, vendas_produto_a, label='Produto A', marker='o') # 'marker' adiciona bolinhas nos pontos
plt.plot(meses, vendas_produto_b, label='Produto B', marker='x')

# Passo 4: Adicionar título e rótulos aos eixos
plt.title('Relatório de Vendas Semestral') # Título do gráfico
plt.xlabel('Mês') # Rótulo do eixo X
plt.ylabel('Quantidade Vendida') # Rótulo do eixo Y

# Passo 5: Adicionar a legenda
plt.legend()

# Passo 6: Exibir o gráfico final
plt.show()
```
